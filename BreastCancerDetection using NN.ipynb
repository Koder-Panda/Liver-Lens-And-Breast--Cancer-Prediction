```python
x = df.drop(columns='label', axis=1)
y=df['label']
 # axis=1=col
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
# standardize the data
'''Standardizing data means transforming it in such a way that it has a mean (average) of 0 and a
standard deviation of 1. This transformation is often referred to as "z-score normalization" or 
"z-score standardization." It is a common preprocessing step in data analysis and machine learning
.The goal of standardizing data is to make it easier to compare and work with data that have 
different scales and units.'''
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.transform(x_test)
```

```python
tf.random.set_seed(3)
#tensorflow.random.set_seed(3) is a way to make sure that random processes in 
# TensorFlow produce the same results every time you run your code. It helps with making your 
# machine learning experiments reproducible
```

'''When you use Keras, you are essentially talking to TensorFlow.
Keras serves as a user-friendly interface or translator that simplifies the interaction 
with the underlying TensorFlow engine.Keras takes care of communicating those 
instructions to TensorFlow, which does the heavy lifting of performing the actual 
computations required for deep learning. This combination makes it easier for 
developers to work with the powerful capabilities of TensorFlow without getting lost in 
its intricacies.'''

 keras.sequential used to stack layers of NN.
 keras.Flatten is used to convert the x_train and y_train to a 1D array.
 no. of neurons= no. of features(here cols).
 keras.Dense means every neuron in this dense layeris connected to the previous layer as well.
 in this model, we have 3 layers-> input,hidden,output layers resp

```python
#Structre of the NN
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),    # no_of features/cols  By adding the comma, 
    #you explicitly indicate that you're creating a tuple with one element,
    keras.layers.Dense(20,activation='relu') ,  # no. of neurons, algo_type
    keras.layers.Dense(2,activation='sigmoid')  # no. of  neurons in output layer=no. of classes(here it is 0 and 1)
])
'''ReLU (Rectified Linear Unit) is like a switch for a light bulb in a dark room. 
If the switch is "on" (positive input), the light shines brightly (output is the same as input). 
If the switch is "off" (zero or negative input), the light is completely off (output is zero). 
It's a way for artificial neurons in a computer to decide whether to be active (light on) or not
(light off) based on the input they receive.
sigmoid=LogisticRegression Curve
'''
```

```python
# Compiling of NN
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy', # use this when dealing with target var no. like 1,2,3..
    # for one hot encoded label variables use categorical_cross_entropy
    metrics=['accuracy'] )
```

```python
# train model
history=model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

```python
#loss is inversely proportional to accuracy

# visualizing loss and acuuracy

#Visualizing accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('MODEL ACCURACY')
plt.xlabel('epoch')
plt.ylabel('Accuracy')
plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
#Visualizing loss
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('MODEL LOSS')
plt.xlabel('epoch')
plt.ylabel('Loss')
plt.legend(['training data', 'validation data'], loc='upper right')
```

```python
# prediction
# by using evaluate func we are predicting as well as comparing(score)
loss,accuracy=model.evaluate(x_test_std,y_test)
# predict x_test values(labels) and compare with y_test values(true label/value)
```
